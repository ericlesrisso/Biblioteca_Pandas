```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
# A função unique no Pandas retorna uma lista dos elementos únicos com base na ocorrência.
vendas.groupby('estado').idcompra.nunique()
```

```python
# A função replace em Python é uma poderosa ferramenta para realizar substituições de palavras em strings
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(" ","_")
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
# Verificando dados nulos
vendas[vendas["estado"].isnull()]
```

```python
# dropna() . Esse método é direto e remove os valores NaN encontrados no DataFrame
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
# Função fillna é usada para preencher os dados nulos no df =, nesse caso, na coluna ESTADO, todos os dados nulos foram susbitituidos por MS
vendas["estado"].fillna("MS",inplace=True)
```

```python
vendas.info()
```

```python
# Função fillna é usada para preencher os dados nulos no df =, nesse caso, na coluna PREÇO, todos os dados nulos foram susbitituidos pela MÉDIA DE PREÇO
media_preco = vendas["Preço"].mean()

vendas["Preço"].fillna(media_preco, inplace=True)
```

```python
media_preco
```

```python
vendas.info()
```

```python
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")

vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
#Filtrando por setor com maior quantidade de vendas
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# Média de preço com frete por nome de departamento
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
vendas_correto.head()
```

```python
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending= False).reset_index()
```

```python
vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
# Quantidade de vendas por mes
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending= False).reset_index()
```

```python
#Inserindo uma nova tabela no projeto
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head()
```

```python
vendas_correto.head()
```

```python
cliente[["idade","renda"]].describe()
```

```python
cliente.info()
```

```python
# Alterando o tipo de dado de uma coluna
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
cliente.dtypes
```

```python
# Fazendo um lef join, para unir os dados da tabela vendas com a tabela cliente
vendas_cliente = vendas_correto.merge(cliente,how= "left", on= "cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idadebandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idadebandeira
```

```python
import matplotlib.pyplot as plt
```

```python
#Plotando um grafico com a biblioteca matplotlib.pyplot

fig = plt.figure(figsize=(7,4))

plt.bar(agg_idadebandeira['bandeira'],agg_idadebandeira['idade'],color='green')

plt.ylabel('Média de idade')

plt.title( 'Idade média por bandeira',loc='center',fontsize=14, fontweight='bold', color='black');
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'],agg_idcanal_renda['renda'],color='blue')

plt.ylabel('Renda média ')

plt.title( 'Renda média por canal de venda',loc='center',fontsize=14, fontweight='bold', color='black');
```

```python
vendas_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(11,4))

plt.xlabel("Data da Venda")
plt.ylabel("Quantidade de Vendas")
x= vendas_por_data['Data']
y = vendas_por_data['idcompra']

plt.plot(x,y, color='Green', linewidth= '2')
plt.show()
```

```python
agg_dept_preco = round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(28,10))

plt.bar(agg_dept_preco['Nome_Departamento'],agg_dept_preco['Preço_com_frete'],color='blue')

plt.ylabel('Preco médio')

plt.title( 'PReço médio por departamento',loc='center',fontsize=14, fontweight='bold', color='black');
```
